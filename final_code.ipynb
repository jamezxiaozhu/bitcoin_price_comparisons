```python
import pandas as pd
from sqlalchemy import create_engine
import datetime 
import time
import numpy as np
```

```python
bit_file = "bitstamp.csv"
bit_df = pd.read_csv(bit_file)
bit_df.dropna(inplace = True)
bit_df.head()
```

```python
coin_file = "coinbase.csv"
coin_df = pd.read_csv(coin_file)
coin_df.dropna(inplace = True)
coin_df.head()
```

```python
gold_file = "gold.csv"
gold_df = pd.read_csv(gold_file)
gold_df.dropna(inplace = True)
gold_df.head()
```

```python
def translate_function_2(raw_datetime):
    return datetime.datetime.strptime(raw_datetime, "%m/%d/%Y").date()
```

```python
datetime.datetime.strptime('1/4/2009', '%m/%d/%Y').date()
```

```python
gold_df['Date'] = gold_df['Date'].apply(translate_function_2)
gold_df.head()
```

```python
new_gold_df= gold_df[['Date','GLD','EUR/USD']]
new_gold_df.head()
```

```python
converted_bit_df = pd.to_datetime(bit_df['Timestamp'], unit='s')
converted_bit_df
```

```python
converted_coin_df = pd.to_datetime(coin_df['Timestamp'], unit='s')
converted_coin_df
```

```python
def translate_function(raw_datetime):
    return datetime.datetime.strptime(str(raw_datetime)[:10], "%Y-%m-%d").date()
```

```python
new_bit_convert = converted_bit_df.apply(translate_function)
print(new_bit_convert)
```

```python
new_coin_convert = converted_coin_df.apply(translate_function)
print(new_coin_convert)
```

```python
bit_df['Timestamp'] = new_bit_convert
new_bit_df = bit_df[['Timestamp','Close']]
new_bit_df.head()
```

```python
new_bit_df.drop_duplicates('Timestamp',inplace=True)
new_bit_df.head()
```

```python
merged_df_gold_bit = pd.merge(new_gold_df,new_bit_df,left_on='Date',right_on='Timestamp')
merged_df_gold_bit
```

```python
coin_df['Timestamp']= new_coin_convert
new_coin_df = coin_df[['Timestamp','Close']]
new_coin_df.head()
```

```python
new_coin_df.drop_duplicates('Timestamp',inplace=True)
```

```python
merged_df_gold_bit_coin = pd.merge(merged_df_gold_bit,new_coin_df,left_on='Date',right_on='Timestamp')
merged_df_gold_bit_coin.head()
```

```python
df_final = merged_df_gold_bit_coin[['Date','GLD','EUR/USD','Close_x','Close_y']]
df_final.head()
```

```python
df_final_renamed = df_final.rename(columns={
    'Date':'Date_Recorded',
    'GLD':'Gold_Price',
    'EUR/USD':'Euros_to_Dollars',
    'Close_x':'Close_Bitstamp',
    'Close_y':'Close_Coinbase'
})
df_final_renamed.head()
```

### Load into Postgres

```python
connection_string = "postgres:password@localhost:5432/bitcoin_comparisons_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
df_final_renamed.to_sql(name='bitcoin_comparisons', con=engine, if_exists='append', index=True)
```
